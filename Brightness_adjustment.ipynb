```python
import os
import cv2

def adjust_brightness(image, factor):
    """Adjust the brightness of an image."""
    return cv2.convertScaleAbs(image, alpha=factor, beta=0)

def process_images_for_brightness(input_folder, output_folder, brightness_factors):
    """Process all images in the input folder, adjust brightness, and save them."""
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('png', 'jpg', 'jpeg')):
            image_path = os.path.join(input_folder, filename)
            image = cv2.imread(image_path)

            if image is None:
                print(f"Failed to load {filename}. Skipping...")
                continue

            base_name = os.path.splitext(filename)[0]

            for idx, factor in enumerate(brightness_factors):
                adjusted_image = adjust_brightness(image, factor)
                output_filename = f"{base_name}_brightness_{idx+1}.jpg"
                output_path = os.path.join(output_folder, output_filename)
                cv2.imwrite(output_path, adjusted_image)
                print(f"Saved: {output_path}")

# Parameters
input_folder = "/content/drive/MyDrive/data/Polythene"  # Replace with the path to your input folder
output_folder = "/content/drive/MyDrive/data/Brightness adjustment"  # Replace with the desired output folder
brightness_factors = [0.2, 0.5, 1.5, 2.0]  # Brightness adjustment factors (0.5 = dimmer, 2.0 = brighter)

# Run the script
process_images_for_brightness(input_folder, output_folder, brightness_factors)
```
